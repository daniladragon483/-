```python
import shlex
from datetime import datetime

class IzmerenieTemperatury:
    def __init__(sam_object, data_tekst, mesto, znachenie):
        sam_object.data = datetime.strptime(data_tekst, "%Y.%m.%d").date()
        sam_object.mesto = mesto
        sam_object.znachenie = float(znachenie)

    def __str__(sam_object):
        return f"Дата: {sam_object.data}\nМесто: {sam_object.mesto}\nЗначение: {sam_object.znachenie}"



def sozdat_objekt(vhodnaya_stroka):
    try:
        elementy_stroki = shlex.split(vhodnaya_stroka)
        
        if len(elementy_stroki) < 4:
            return None

        noviy_objekt = IzmerenieTemperatury(elementy_stroki[1], elementy_stroki[2], elementy_stroki[3])
        return noviy_objekt

    except Exception:
        return None

testovaya_stroka = 'TipObekta 2023.10.25 "улица Пушкина" 22.3'
rezultat = sozdat_objekt(testovaya_stroka)

if rezultat:
    print(rezultat)


```

```python
import shlex
import tkinter as tk
from tkinter import ttk, messagebox
from datetime import datetime

class TemperatureMeasurement:
    def __init__(self, date_str, location, value):
        self.date = datetime.strptime(date_str, "%Y.%m.%d").date()
        self.location = location
        self.value = float(value)

def parse_line(line):
    try:
        parts = shlex.split(line)
        if len(parts) >= 4:
            return TemperatureMeasurement(parts[1], parts[2], parts[3])
    except:
        pass
    return None

def load_from_file(filename):
    data = []
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            for line in f:
                obj = parse_line(line.strip())
                if obj:
                    data.append(obj)
    except:
        print("Файл не найден")
    return data

class App:
    def __init__(self, root, data_list):
        self.root = root
        self.items = data_list

        self.tree = ttk.Treeview(self.root, columns=("1", "2", "3"), show="headings")
        self.tree.pack()

        self.e1 = tk.Entry(self.root)
        self.e1.pack()
        self.e2 = tk.Entry(self.root)
        self.e2.pack()
        self.e3 = tk.Entry(self.root)
        self.e3.pack()

        tk.Button(self.root, text="Добавить", command=self.add).pack()
        tk.Button(self.root, text="Удалить", command=self.delete).pack()

        self.show()

    def show(self):
        for i in self.tree.get_children():
            self.tree.delete(i)
        for obj in self.items:
            self.tree.insert("", "end", values=(obj.date, obj.location, obj.value))

    def add(self):
        color = self.e4.get().strip().lower()
        if color not in ["красный", "синий", "белый"]:
            messagebox.showwarning("ошибка", "допустимые цвета: красный, синий, белый")
            return
        raw = f'Entry {self.e1.get()} "{self.e2.get()}" {self.e3.get()}'
        new_obj = parse_line(raw)
        if new_obj:
            self.items.append(new_obj)
            self.show()

    def delete(self):
        sel = self.tree.selection()
        if sel:
            idx = self.tree.index(sel[0])
            del self.items[idx]
            self.show()

my_data = load_from_file("data.txt")
root = tk.Tk()
main_app = App(root, my_data)
root.mainloop()
```
