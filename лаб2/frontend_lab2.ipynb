```python
import tkinter as tk
from tkinter import ttk, messagebox
from data_manager import parse_line, load_from_file

class App:
    def __init__(self, root, data_list):
        self.root = root
        self.items = data_list

        self.tree = ttk.Treeview(self.root, columns=("1", "2", "3", "4"), show="headings")
        self.tree.heading("1", text="Дата")
        self.tree.heading("2", text="Локация")
        self.tree.heading("3", text="Значение")
        self.tree.heading("4", text="RGB")
        self.tree.pack()

        self.e1 = tk.Entry(self.root)
        self.e1.pack()
        self.e2 = tk.Entry(self.root)
        self.e2.pack()
        self.e3 = tk.Entry(self.root)
        self.e3.pack()
        self.e4 = tk.Entry(self.root)
        self.e4.pack()

        tk.Button(self.root, text="Добавить", command=self.add).pack()
        tk.Button(self.root, text="Удалить", command=self.delete).pack()

        self.show()

    def show(self):
        for i in self.tree.get_children():
            self.tree.delete(i)
        for obj in self.items:
            self.tree.insert("", "end", values=(obj.date, obj.location, obj.value, obj.rgb))

    def add(self):
        color = self.e4.get().strip().lower()
        if color not in ["красный", "синий", "белый"]:
            messagebox.showwarning("ошибка", "допустимые цвета: красный, синий, белый")
            return
        raw = f'Entry {self.e1.get()} "{self.e2.get()}" {self.e3.get()} {self.e4.get()}'
        
        new_obj = parse_line(raw) 
        
        if new_obj:
            self.items.append(new_obj)
            self.show()

    def delete(self):
        sel = self.tree.selection()
        if sel:
            idx = self.tree.index(sel[0])
            del self.items[idx]
            self.show()

if __name__ == "__main__":
    my_data = load_from_file("data.txt")
    root = tk.Tk()
    main_app = App(root, my_data)
    root.mainloop()
```
